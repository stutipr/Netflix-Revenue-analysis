# Netflix revenue over time

```python
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Global Revenue'])
plt.title('Netflix Global Revenue over time')
plt.xlabel('Date')
plt.ylabel('Global Revenue')
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,.2f}".format(x)))
```

```python
avg_user=df.groupby('Year')['Netflix Global Users'].mean()

plt.figure(figsize=(10,6))
plt.plot(avg_user.index,avg_user.values)
plt.title('Avg Netflix Global User over years')
plt.xlabel('Year')
plt.ylabel('Netflix Global users')
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,.2f}".format(x)))
```

```python
df_clean=df.dropna(subset=['UACN Revenue','EMEA Revenue','LATM Revenue','APAC Revenue'])

region_revenue=df_clean[['UACN Revenue','EMEA Revenue','LATM Revenue','APAC Revenue']].sum()
```

```python
plt.figure(figsize=(8,6))
region_revenue.plot(kind='bar',color='blue')
plt.title('Netflix Regional Revenue over time')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,.2f}".format(x)))
```

# Domestic vs International Revenue visualization


```python
import matplotlib.pyplot as plt

# Calculate average revenue per year
avg_domestic_revenue = df.groupby('Year')['Domestic Revenue'].mean()
avg_international_revenue = df.groupby('Year')['International Revenue'].mean()

# Set the positions of the bars on the x-axis
x = avg_international_revenue.index

# Plot the bar graphs
plt.bar(x, avg_domestic_revenue, width=0.4, align='center', label='Domestic Revenue', color='green')
plt.bar(x, avg_international_revenue, width=0.4, align='edge', label='International Revenue', color='yellow')

# Customize the graph
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Domestic & International Revenue over the Years')
plt.legend()
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,.2f}".format(x)))
# Display the graph
plt.show()
```

# Revenue wise region visualization

```python
 fig,axs = plt.subplots(2,2)
fig.suptitle('UACN vs EMEA vs LATM vs APAC Revenue Trends',fontdict ={'family':'fantasy','color':'green'}, fontsize =15)
axs[0,0].plot(df['Date'],df['UACN Revenue'],color= "yellow")
axs[0,1].plot(df['Date'],df['EMEA Revenue'],color= "blue")
axs[1,0].plot(df['Date'],df['LATM Revenue'],color= "red")
axs[1,1].plot(df['Date'],df['APAC Revenue'],color= "magenta")
axs[0,0].set_title('UACN Revenue', fontdict= {'family':'fantasy','color':'black','size':12})
axs[0,1].set_title('EMEA Revenue',fontdict= {'family':'fantasy','color':'black','size':12})
axs[1,0].set_title('LATM Revenue', fontdict= {'family':'fantasy','color':'black','size':12})
axs[1,1].set_title('APAC Revenue', fontdict= {'family':'fantasy','color':'black','size':12})

for ax in axs.flat:
    ax.set(xlabel='Time', ylabel='Revenue' )
for ax in axs.flat:
    ax.set_xticklabels([])
    ax.label_outer()
```
